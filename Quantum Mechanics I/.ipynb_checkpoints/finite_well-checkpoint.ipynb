```python
import math
from matplotlib import pyplot
import numpy as np

m=0.06
L=90
u=240
h=6.851734267*(10**52)



def lhs(E):
    return math.tan((L*math.sqrt(2*m*(E+u)))/2*h)

def rhs(E):
    return math.sqrt((-E)/(-u+E))
    
    
E_list_lhs = []
y_list_lhs = []
E_list_rhs = []
y_list_rhs = []


for energy in range (1,20):
        E_list_lhs.append(lhs(energy))
        E_list_rhs.append(rhs(energy))
        y_list_lhs.append(energy)
        y_list_rhs.append(energy)
        if lhs(energy)==rhs(energy):
            print (energy)


pyplot.plot (E_list_lhs, y_list_lhs)
pyplot.plot (E_list_rhs, y_list_rhs)
pyplot.yscale('log')
pyplot.show()
    
pyplot.plot (E_list_lhs, y_list_lhs)
pyplot.plot (E_list_rhs, y_list_rhs)
pyplot.show()
```
